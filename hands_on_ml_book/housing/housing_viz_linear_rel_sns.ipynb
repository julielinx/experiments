```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
housing_raw = pd.read_csv('https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv')
housing_raw.head()
```

```python
sns.pairplot(housing_raw, plot_kws={'linewidth':0, 'size':15, 'alpha':0.3})
```

```python
sns.pairplot(housing_raw, kind='reg', diag_kind='kde')
```

```python
sns.lmplot(x='longitude', y='latitude', data=housing_raw)
```

```python
sns.lmplot(x='longitude', y='latitude', data=housing_raw, robust=True)
```
