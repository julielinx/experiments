```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
housing_raw = pd.read_csv('https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv')
```

```python
housing_raw.head()
```

```python
plt.figure(figsize=(8,10))
sns.scatterplot(x=housing_raw['longitude'], y=housing_raw['latitude'], alpha=0.5, linewidth=0)
```

```python
plt.figure(figsize=(8,10))
sns.relplot(x='longitude', y='latitude',
            data=housing_raw,
            alpha=0.4,
            hue='ocean_proximity',
#             style='ocean_proximity',
            height=10, aspect=.8,
           linewidth=0, size=40)
```

```python
plt.figure(figsize=(8,10))
sns.relplot(x='longitude', y='latitude',
            data=housing_raw,
            alpha=0.4,
            hue='housing_median_age',
            palette='RdYlGn',
            col='ocean_proximity',
            col_wrap=2,
#             height=10, aspect=.8,
           linewidth=0, size=40)
```

```python
plt.figure(figsize=(8, 10))
sns.regplot(x=housing_raw['longitude'], y=housing_raw['latitude'], fit_reg=False)
```

```python
plt.figure(figsize=(8, 10))
sns.regplot(x=housing_raw['longitude'], y=housing_raw['latitude'], fit_reg=False, scatter_kws={'alpha':0.5})
```

```python
plt.figure(figsize=(8, 10))
sns.regplot(x=housing_raw['longitude'], y=housing_raw['latitude'], fit_reg=False, scatter_kws={'alpha':0.3, 's':25, 'color':'darkred'})
```

```python
sns.regplot(x=housing_raw['households'], y=housing_raw['total_rooms'], line_kws={'color':'darkred', 'lw':3,'alpha': 0.7}, scatter_kws={'s':40, 'alpha':0.2})
```

```python
sns.regplot(x=housing_raw['households'], y=housing_raw['total_rooms'], line_kws={'color':'darkred', 'lw':3,'alpha': 0.7}, scatter_kws={'s':40, 'alpha':0.2})
```

```python
sns.lmplot(x='longitude', y='latitude', data=housing_raw, hue='ocean_proximity', palette='Set1', fit_reg=False, scatter_kws={'alpha':0.3, 's':25})
```
